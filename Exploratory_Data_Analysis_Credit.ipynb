```python
#Separating continuous and categorical columns from the dataset
for i in K.columns:
    if(K[i].dtypes == 'object'):
        sb.boxplot(K[i],K.Balance)#Boxplot between categorical and continuous columns
        plt.xlabel(i)
        plt.ylabel('Balance')
        plt.show()
    else:
        plt.scatter(K[i],K.Balance)# Scatter plot between continuous and continuous columns
        plt.xlabel('Balance')
        plt.ylabel(i)
        plt.show()
```

```python
for i in K.columns:
   if(K[i].dtypes=="object"):
    sb.countplot(K.Married,hue=K[i])#Countplot with hue between categorical and categorical columns
    plt.ylabel(i)
    plt.show()
```
